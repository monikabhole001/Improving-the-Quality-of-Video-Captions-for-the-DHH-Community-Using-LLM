```python
from pathlib import Path # Library for handling filesystem paths
from youtube_transcript_api import YouTubeTranscriptApi as yta
import pandas as pd
```

```python
# Initialize the base directory
base_dir = Path('videos')
```

```python
def list_files_and_folders(base_dir, files_list):
    """
    Recursively lists all files and folders starting from the base directory.
    """
    for item in base_dir.iterdir():
        if item.is_dir():
            # It's a directory, print directory name and explore it recursively
#             print(f"Directory: {item}")
            list_files_and_folders(item, files_list)
        else:
            # It's a file, just print the file name
#             print(f"File: {item}")
            files_list.append(item)
```

```python
all_files = []
list_files_and_folders(base_dir, all_files)
```

```python
df_ds = pd.read_csv("DHH_data_source.csv", encoding='ISO-8859-1')
```

```python
df_ds
```

```python
df_groundtruth= pd.read_csv("groundtruth-caption.csv")
```

```python
df_groundtruth
```

```python
label = []
category = []
for i,row in df_groundtruth.iterrows():
    label.append(int(Path(row[0]).stem)) 
    category.append(Path(row[0]).parent.stem)
```

```python
df_groundtruth["youtube_caption"] = [""]*df_groundtruth.shape[0]
```

```python
df_groundtruth["Label"]= label
```

```python
df_groundtruth["Category"]= category
```

```python
df_groundtruth 
```

```python
data = []
# Now, all_files contains all file paths
for file in all_files:
#     print(file.parent.stem)
    try:
        vid = df_ds[(df_ds["Label"] == int(file.stem)) & (df_ds["Category"] == file.parent.stem)]["Video_Id"].values[0]
#         print(vid, type(vid))
        tx = yta.get_transcript(vid, languages=['de', 'en' ])

        outls = []
        for i in tx:
            outtxt = i['text']
            outls.append(outtxt)

        caption = " ".join(outls)
        indexer = (df_groundtruth["Label"] == int(file.stem)) & (df_groundtruth["Category"] == file.parent.stem)
        df_groundtruth.loc[indexer, "youtube_caption"] = caption
    except Exception as e:
        print(e)
#         df_groundtruth[df_groundtruth["Label"] == int(file.stem)]["youtube_caption"] = "caption not found"
        indexer = df_groundtruth["Label"] == int(file.stem)
        df_groundtruth.loc[indexer, "youtube_caption"] = "caption not found"
```
